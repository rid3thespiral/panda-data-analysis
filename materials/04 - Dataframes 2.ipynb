## Metodi where() e query() 
- metodi utilizzati per operazioni di *filtering* di un dataset

```python
import pandas as pd
titanic = pd.read_csv("titanic.csv")
titanic.head(10)
```

Voglio estrarre tutti passeggeri che erano in prima classe (Pclass = 1)

```python
# gia' visto: estrazione / filtering in base ad una condizione

mask = titanic["Pclass"] == 1
titanic[mask]
```

**where()**

- mi restituisce il dataframe originale con "NaN" dove la condizione non e' vera

```python
titanic.where(titanic["Pclass"] == 1)
```

*tramite il parametro "other" posso visualizzare un risultato diverso da "NaN" quando la condizione non e' vera*

```python
titanic.where(titanic["Pclass"] == 1, other = "FALSO" ) # inplace = False
```

**query()**


```python
# l'espressione deve essere una stringa per essere valutata

titanic.query("Pclass == 1") # attenzione all'uso delle virgolette
```

## metodo agg()
- lo utilizziamo per crearci delle statistiche custom del nostro dataset

```python
import pandas as pd
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.describe(include = "all")
```

```python
titanic.agg("max") # stesso risultato con titanic.max()

# axis = 0 oppure axis = "index": applica la funzione su ogni colonna
# axis = 1 oppure axis = 'columns': applica la funzione ad ogni riga
```

```python
titanic.agg(["max", "min"])
```

```python
# tramite l'uso di un dizionario possiamo scegliere quale statistica applicare e a quale colonna
titanic.agg({"Fare": "max", "Age": ["max", "min", "mean"]})
```

## Metodo copy()
- crea una copia esatta di un oggetto in Pandas e la salva in una nuova zona di memoria completamente separata dall'oggetto originale

```python
import pandas as pd
titanic = pd.read_csv("titanic.csv")
```

Supponiamo di voler fare alcune manipolazioni sulla nostra Series o Dataframe ma senza toccare la Series o dataframe originale

```python
# assegno la mia Series relativa all'eta' ad una variabile che chiamo "age"
age = titanic["Age"]
age
```

```python
# mi aspetto che le operazioni sulla variabile "age" siano "isolate", non riguardino l'oggetto originale
age[0] = 1000
```

```python
age
```

```python
titanic["Age"]
```

```python
titanic
```

con il metodo copy() ottengo una copia completamente disgiunta dall'oggetto originale

```python
prezzo = titanic["Fare"].copy()
prezzo
```

```python
titanic["Fare"]
# La Series originale non e' stata modificata!
```

## Operazioni con le stringhe 1
- Quasi tutti i metodi built-in in python che permettono operazioni sulle stringhe hanno un corrispettivo in Pandas
- La differenza sintattica sostanziale e' che in Pandas bisogna usare l'attributo **.str** prima di usare un metodo sulle stringhe 

```python
import pandas as pd
titanic = pd.read_csv("titanic.csv")
```

```python
nomi = titanic["Name"].head(10).copy()
nomi
```

```python
stringa = "Mi chiamo Dario"
stringa.upper()
```

```python
nomi.str.upper() # in Pandas occore utilizzare l'attributo .str prima di applicare un metodo sulle stringhe
```

### upper(), lower(), title(), len(), contains()

```python
nomi.str.upper() # tutte le lettere della stringa vengono rese maiuscole
```

```python
nomi.str.lower() # # tutte le lettere della stringa vengono rese minuscole
```

```python
nomi.str.title() # converte il primo carrattere di ogni parola in maiuscola
```

```python
nomi.str.len() # conta il numero di caratteri di ogni stringa inclusi gli spazi tra ogni parola
```

```python
# posso usare ".contains() "per fare operazioni di filtering (ritorna una Series di booleani)
nomi[nomi.str.contains("Mr")]
```

Buona pratica e' normalizzare prima le stringhe su cui si lavora e poi applicare un metodo.

Per normalizzare intendo formattare tutte le stringhe in modo che siano "uguali": ad esempio tutte minuscole ( o maiuscole)

```python
nomi.str.lower().str.contains("mrs")
```

Se non voglio "normalizzare", devo ricordarmi del parametro "case"

```python
nomi.str.contains("mr", case = False)
```

## Operazioni con le stringhe 2

### startswith(), endswith()

```python
import pandas as pd
titanic = pd.read_csv("titanic.csv")
nomi = titanic["Name"].head(10).copy()
nomi
```

```python
# lo utlizzo per verificare se le stringhe all'interno della Series iniziano con la lettera A
nomi.str.startswith("A") # ritorna una series con valori booleani (True se ogni stringa inizia con la letter "A")
```

```python
nomi.str.endswith("J")
```

### .replace()

```python
nomi.str.replace("Mr.", "Mister") # attenzione al parametro "regex = True" di default
# con regex = True il pattern "Mr." significa sostituisci l'espressione "Mr" + "qualsiasi altro carattere" con "Mister"
```

```python
nomi.str.replace("Mr.", "Mister", regex = False)
# regex = False fa si che in questo caso sostuisco esattamente l'espressione "Mr." 
# cioe' la notazione "." che nelle espressioni regex rappresenta "qualsiasi altro carattere tranne quello di nuova riga",
# ora rappresenta solo il punto (.)
```

## Operazioni con le stringhe 3

### split() e get()

```python
import pandas as pd
titanic = pd.read_csv("titanic.csv")
nomi = titanic["Name"].head(10).copy()
nomi
```

Il metodo split mi divide una stringa in base ad un separatore

```python
nomi.str.split(",") # Passo tra parentesi il "separatore", il carattere che voglio usare per separare la mia stringa
nomi.str.split(",").str.get(0) # get per estarre elementi da una lista
```

```python
# parametro "n" per stabilire a priori il numero di split
nomi.str.split(" ", n = 1)
```

```python
# parametro expand: assegna ogni elemento del mio split ad una nuova colonna
nomi.str.split(",", expand = True)
```

```python
titanic[["Cognome", "Nome"]] = nomi.str.split(",", expand = True)
```

```python
titanic.head(10)
```

##  Indice Gerarchico (Multi-Indexing) parte 1
- L'indice incorpora livelli multipli (di indici): "e' un indice che contiene piu' indici"
- più livelli di indicizzazione delle mie informazioni
- Utilizzato per immagazzinare dati ad alta dimensione ("2D+") in una tabella a due dimensioni

```python
import pandas as pd
nba = pd.read_excel("nba.xlsx")
nba
```

Avevamo gia' visto come settare un indice singolo

```python
nba.set_index("POS")
```

Per settare un indice multiplo basta passare al parametro "keys" una lista di nomi di colonne da usare come indici

```python
nba.set_index(keys = ["POS", "TEAM"], inplace = True)

# Buona norma avere l'indice "ordinato"
nba.sort_index(inplace = True)
nba.head(50)
```

```python
nba.index
```

```python
type(nba.index[0]) # e'una tupla
```

Otteniamo i valori dell'indice ad un dato livello tramite get_level_values()

```python
nba.index.get_level_values(1)
```

Possiamo cambiare il livello di un indice direttamente tramite il metodo swaplevel()

#### L'indice piu' esterno in generale dovrebbe avere meno valori unici rispetto all'indice piu' interno

```python
nba.swaplevel()
```

##  Indice Gerarchico (Multi-Indexing) parte 2

```python
import pandas as pd
nba = pd.read_excel("nba.xlsx", index_col = [2,1]).sort_index() # index_col : int, o lista di interi (posizioni delle colonne)
nba
```

#### Estrazione di righe e colonne quando abbiamo un Multi-Index

```python
# voglio visualizzare tutti i giocatori che giocano nella posizione di Centro ("C")
nba.loc["C"]
```

Voglio estrarre tutti i giocatori di Atlanta ("Atl") che giocano come "Centro" ("C")
- livello 0 del Multi-Index: seleziono "C"
- livello 1 del Multi-Index: seleziono "Atl"

Quando lavoro con i Multi-Index devo passare una **tupla** invece di una lista

```python
nba.loc[("C", "Atl")]
```

Voglio estrarre tutti i giocatori di Atlanta ("Atl") che giocano come "Centro" ("C") ma conoscere solo il loro nome
- livello 0 del Multi-Index: seleziono "C"
- livelli 1 del Multi-Index seelziono "Atl"
- Devo selezionare la colonna "FULLNAME"

```python
nba.loc[("C", "Atl"), "FULLNAME"]
```

Voglio estrarre tutti i giocatori di Atlanta ("Atl") che giocano come "Centro" ("C") e conoscere nome e gare giocate
- livello 0 del Multi-Index: seleziono "C"
- livelli 1 del Multi-Index seleziono "Atl"
- Devo selezionare la colonna "FULLNAME" e "GP" all'interno di una lista

```python
nba.loc[("C", "Atl"), ["FULLNAME", "GP"]]
```

Voglio estrarre tutti i giocatori di Atlanta ("Atl") che giocano come "Centro" ("C") e Guardia ("G")
- livello 0 del Multi-Index: seleziono "C" e "G"
- livello 1 del Multi-Index: seleziono "Atl"

```python
# nba.loc[(["C", "G"], "Atl")] # Pandas vuole che specifichi che vuoi tutte le colonne (:)
nba.loc[(["C", "G"], "Atl"), :]
```

Vogliamo ottenere tutti i giocatori che giocano ad Atlanta ("Atl")
- livello 0 Multi-Index: nessuno slicing (voglio tutti i ruoli)
- livello 1 Multi-Index: selezionare solo "Atl"

```python
# posso pensare di scrivere:
#nba.loc[(:, "Atl"), :]
```

Occorre utlizzare il comando **slice**

```python
nba.loc[(slice(None), slice("Atl")), :]
```

Oppure utilizzare la classe IndexSlice che accetta una sintassi simile a quella vista in caso di indice singolo

```python
nba.loc[pd.IndexSlice[:, "Atl"], :]
```
