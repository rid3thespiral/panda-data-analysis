### metodo transpose()

## stack() e unstack()

```python
import pandas as pd
titanic = pd.read_csv("titanic.csv", index_col = "Name", usecols = ["Name", "Pclass", "Sex", "Age"])

titanic
```

**stack()**
- Il metodo stack prende tutti nomi delle colonne e li rimodella verticalmente creando un indice unico
- Otteniamo una Series con un Multindex

```python
titanic.stack(dropna = False)

# Series con Multi-index
```

*Il metodo stack in pratica prende tutti i nomi delle colonne e li traspone, in modo che diventino l'indice piu' interno (livello 1) di un multi-index*

#### unstack()

L'indice piu' interno viene trasposto e diventa una colonna

```python
titanic = pd.read_csv("titanic.csv", index_col = ["Pclass", "Name"], usecols = ["Name", "Pclass", "Sex", "Age"]).head()
titanic
```

```python
titanic.unstack(level = 0)

# Creo una multiindex per le colonne
```

## melt()

```python
import pandas as pd
pop = pd.read_excel("pop.xlsx")
pop
```

Il metodo melt() fa una conversione del dataframe da "wide to long" (da largo a lungo)
- e' utile quando si hanno piu' colonne che contengono informazioni simili ( ad esempio vendite per mese, popolazione per ogni anno, etc..): queste sono chiamate **value_vars** (variabili misurabili): queste sono le colonne che vogliamo spostare da "orizzontale a verticale" da "wide a long"
- una (o piu') colonne che fungono da indentificatore **id_vars**: rappresenta la colonna che vogliamo preservare all'interno del dataframe

```python
pop.melt(id_vars = "Country Name", value_vars = [2016, 2017, 2018, 2019])
# le colonne relative agli anni vengono condensate in una unica colonna chiamata "variable" di default
```

```python
pop.melt(id_vars = "Country Name",
        var_name="anno",
        value_name='popolazione')
```

## pivot()

```python
import pandas as pd
df = pd.read_csv("salaries.csv")
df
```

Il metodo pivot fa il contrario di melt : si passa da lungo a largo (long to wide)

```python
df.pivot(index = "Season", columns = "Player", values = "Salary").fillna("non ha giocato")
```

## groupby() parte 2

```python
import pandas as pd
titanic = pd.read_csv("titanic.csv")
titanic.head()
```

#### groupby() in combinazione con il metodo agg()

- Troviamo l'eta' media, massima e minima dei passeggeri in base al sesso

```python
# step 1: SPLIT, creo il mio oggetto groupby 

titanic.groupby(by = "Sex")
```

```python
# step 2 e 3: APPLY e COMBINE: applico una funzione e combino i risultati

titanic.groupby("Sex").agg({"Age": ["mean", "max", "min"]})
```

- Troviamo l'eta' media, massima e minima dei passeggeri in base al sesso e il costo del biglietto medio

```python
titanic.groupby("Sex").agg({"Age": ["mean", "max", "min"], "Fare": "mean"})
```

- Come esempio sopra ma in base alla classe della cabina (Pclass)
-- devo creare un oggetto groupby con multiindex

```python
titanic.groupby(by = ["Sex", "Pclass"])
```

```python
titanic.groupby(["Sex", "Pclass"]).agg({"Age": ["mean", "max", "min"], "Fare": "mean"})
```
