## Broadcasting

- Concetto preso dalla libreria Numpy
- Le operazioni base (+, -, *, ...) sono *elementwise*
- permette di eseguire operazioni *elementwise*, anche se le series / dataframe hanno dimensione diversa --> trasforma series / daatframe in modo che abbiano tutti la stessa dimensione (questa operazione e' detta **broadcasting**)

```python
import pandas as pd

# somma tra series con la stessa dimensione

s1 = pd.Series(data = [1,2,3])
s2 = pd.Series(data = [10,20,30])

s1 + s2
```

```python
# somma tra series con dimensione differente
# ottengo una series di dimensione pari a quella della series di dimensione maggiore senza darmi un errore
s3 = pd.Series(data = [1,2,3,4])
s1 + s3
```

```python
# somma tra dataframes con la stessa dimensione
df1 = pd.DataFrame(data = [[1,2], [3,4]], columns=['a', 'b'])
df2 = pd.DataFrame(data = [[5,6], [7,8]], columns=['a', 'b'])
```

```python
df1 + df2
```

```python
# somma tra dataframes con  dimensione differente
df3 = pd.DataFrame(data = [[1,2,3], [1,2,3], [1,2,3]], columns=['a', 'b', 'c'])
df2 + df3
```

Vediamo meglio cosa significa che le **operazioni sono elementwise**

```python
nba = pd.read_excel("nba.xlsx").head()
nba
```

```python
nba["GP"] + 5 # l'operazione "+5" viene propagata a tutti gli elementi della colonna
```

```python
nba[["GP", "MPG"]] * 10
```

posso crearmi anche una nuova colonna in questo modo

```python
# creo colonna che chiamo "TOT" che mi da la somma di punti, rimbalzi e assist a partita

nba["TOT"] = nba["PPG"] + nba["RPG"] + nba["APG"]
nba

# "numero" + NaN = NaN
```

## Estrazione in base a piu' condizioni

```python
import pandas as pd
nba = pd.read_excel("nba.xlsx", index_col = "FULLNAME")
nba
```

#### Esempio 1: Proviamo ad estrarre i giocatori che hanno media punti superiore a 20 ed eta' inferiore a 25
- dovro' creare due maschere, una per ogni condizione

```python
mask1 = nba["PPG"] > 20
mask1
```

```python
mask2 = nba["AGE"] < 25
mask2
```

```python
# voglio che entrambe le condizioni siano soddisfatte, quindi uso operatore "&" (diverso da operatore logico "and")

nba[mask1 & mask2]
```

 estraggo i giocatori che hanno media punti superiore a 20 ed eta' inferiore a 25, ma voglio solo sapere in quale squadra giocano

```python
nba[mask1 & mask2]["TEAM"]
```

```python
# oppure tramite operatore loc[]
nba.loc[mask1 & mask2, "TEAM"]
```

#### Esempio 2: Proviamo ad estarre i giocatori che giocano nei Lakers (Lal) con eta' superiore a 30


```python
mask1 = nba["TEAM"] == "Lal"
mask2 = nba["AGE"] > 30
```

```python
nba[mask1 & mask2]
```

```python
# perche' non usare operatore "and"?
nba[mask1 and mask2]

```

```python
mask2
# "and" lavora sul contenuto della series nella sua totalita'
# la series mask1 e' True? o e' False? al suo interno contiene elementi che sono alcuni True e alcuni False
# Occorre usare l'operatore "bitwise": voglio fare confronto elemento per elemento
```

#### Esempio 3: Proviamo ad estarre giocatori che giocano nei Lakers (Lal) OPPURE  nei Boston Celtics (Bos)

```python
mask1 = nba["TEAM"] == "Lal"
mask2 = nba["TEAM"] == "Bos"

nba[mask1 | mask2]
```

#### Esempio 4: Proviamo ad estarre giocatori che giocano nei Lakers (Lal) OPPURE  nei Boston Celtics (Bos), ma che hanno segnato almeno 20 punti a partita di media

```python
# creo tre maschere
mask1 = nba["TEAM"] == "Lal"
mask2 = nba["TEAM"] == "Bos"
mask3 = nba["PPG"] > 20

# attenzione all'uso delle parentesi

nba[(mask1 | mask2) & mask3]
```

#### Esempio 5: Come esempio 4, ma voglio conoscere solo la posizione e l'eta'

```python
mask1 = nba["TEAM"] == "Lal"
mask2 = nba["TEAM"] == "Bos"
mask3 = nba["PPG"] > 20

# attenzione all'uso delle parentesi

nba[(mask1 | mask2) & mask3][["POS", "AGE"]]
```

```python
# oppure con operatore loc[]

nba.loc[(mask1 | mask2) & mask3, ["POS", "AGE"]]
```

## Il metodo apply()
- Utilizzato per applicare una funzione ad ogni elemento di una series, quindi ha come parametro una funzione: .apply(func)

```python
import pandas as pd
nba = pd.read_excel("nba.xlsx")
nba
```

#### Esempio 1: contiamo quante lettere formano i nomi dei giocatori

```python
# Applico, tramte metodo apply(), la fuznione built-in di python "len" ad ogni elemento della series "FULLNAME"
nba["FULLNAME"].apply(len)
```

```python
# posso creare una nuova colonna
nba["num_lettere"] = nba["FULLNAME"].apply(len)
nba
```

#### Esempio 2: definisco io una funzione che voglio applicare e la passo al metodo apply()

Definiamo una funzione che in base all'eta' del giocatore mi restituisce i seguenti aggettivi:
    - vecchio se AGE > 33
    - giovane se 23 < AGE <= 33
    - rookie se AGE <= 23

```python
def age(numero):
    if numero > 33:
        return "vecchio"
    elif numero <= 23:
        return "rookie"
    else:
        return "giovane"
```

```python
nba["AGE"].apply(age)
```

#### Esempio 3: tramite funzioni anonime (lambda)
voglio arrotondare l'eta' dei giocatori a zero decimali

```python
nba["AGE"].apply(lambda x:round(x,0))
```

#### metodo apply() "lungo le colonne (da sinistra a destra)": attenzione al parametro axis

```python
# seleziona una unica osservazione
nba.loc[0]
```

```python
# posso accedere ad ogni elemento della osservazione, ad esempio AGE si trova in "posizione 3"
nba.loc[0][3]
```

```python
def titolare(riga):
    age = riga[3]
    partite = riga[4]
    minuti = riga[5]
    
    if age < 25 and partite > 50 and minuti > 25:
        return "giovane titolare"
    else:
        return "non so"
```

```python
nba["titolare"] = nba.apply(titolare, axis = 1) # axis = 1 perche' mi sto muovendo lungo le colonne (orizzontalmente)
nba
```

## I metodi map() e applymap()

 

```python
import pandas as pd
nba = pd.read_excel("nba.xlsx")
nba
```

### map ()
- e' un metodo per le Series
- Usato principalmente per sostituire (mappare) ogni valore della Series con un altro valore

```python
nba["POS"] = nba["POS"].map({"C": "Centro",  "C-F": "Ala Grande", "F-C": "Ala Grande", "F": "Ala piccola","G": "Guardia"})
nba
```

### metodo applymap()
- Utlizzato sui dataframe
- applica una funzione sul dataframe elemento per elemento

```python
nba.applymap(lambda x: len(str(x))) # ritorna un dataframe
```

## Il metodo astype()
- Serve per convertire un oggetto di pandas in un *dtype* qualsiasi

- Nel caso di conversione in numeri, funziona solo se non ci sono NaN

```python
import pandas as pd
nba = pd.read_excel("nba.xlsx")
nba
```

```python
nba.dtypes
```

```python
nba.info()
```

```python
# "astype" non ha il parametro inplace
nba["AGE"] = nba["AGE"].astype("int8")
nba.info()
```

*conversione in un numero non funziona se ci sono NaN*

```python
#nba["APG"].astype("int")
nba["APG"].astype("object")
```

```python
nba = pd.read_excel("nba.xlsx")
nba.info()
```
